```python
import dask.dataframe as dd
import dask
import pandas as pd
#from  dask import dask-expr as dde
```

```python
df = pd.read_csv("./data/raw/twitter/cyberbullying_tweets.csv")
```

```python
df.shape  # dask dask.compute(df.shape)
```

```python
df.sample(n=50)
```

```python
df = df.rename(columns={"tweet_text": "text"})
```

```python
df.head(10)
```

```python
unique_split_names = list(dask.compute(df["cyberbullying_type"].unique())[0])
unique_split_names
```

```python
def func(row:dict[str : str])-> int:
    try:
        if row['cyberbullying_type'] == 'not_cyberbullying':
            val = 0
        else:
            val = 1
        return  val
    except:
        return -1
df['label'] = df.apply(func, axis=1, meta=float)
```

```python
df.head(10)
```

```python
unique_label_values = list(dask.compute(df["label"].unique())[0])
```
